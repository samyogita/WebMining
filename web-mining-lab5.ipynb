```python
import networkx as nx
import pylab as plt
D=nx.DiGraph()
D.add_weighted_edges_from([('A','B',1),('A','C',1),('C','A',1),('B','C',1),('B','D',1),('A','D',1),('C','D',1)])
nx.draw(D,with_labels=True)
plt.show()
print(nx.pagerank(D))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
D = nx.DiGraph()
D.add_edges_from([('A','B'),('A','C'),('C','A'),('B','C'),('B','D'),('A','D'),('C','D'),('C','B')])
plt.figure(figsize =(10,5))
nx.draw_networkx(D, with_labels = True)
hub, authority = nx.hits(D, max_iter = 100, normalized = True)
print("Hub: ", hub)
print()
print("Authority: ", authority)
```
