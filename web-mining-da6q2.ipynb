### Create a Python programme that uses the K-means clustering algorithm and displays all clusters in different colours.

### Importing necessary modules

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

### Loading the dataset 

```python
iris = load_iris()

X = iris.data
y = iris.target
```

### Plotting the data

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='summer')
plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

### Training the model and fitting the data

```python
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

### Finding the centers

```python
centers = km.cluster_centers_
print(centers)
```

```python
new_labels = km.labels_
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='Pastel2', edgecolor='navy', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='Pastel1', edgecolor='navy', s=150)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[0].tick_params(direction='in', length=10, width=5, colors='navy', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='navy', labelsize=20)
axes[0].set_title('Actual value')
axes[1].set_title('Predicted value');
```
